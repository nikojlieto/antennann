```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
githuburl = 'https://github.com/nikojlieto/antennann/raw/main/localization%20XLP.csv'
#githuburl.encode('utf-8').strip()
df = pd.read_csv(githuburl, dtype={'epc96':"category", 'antennastate':"category"})
```

```python
df['epc96'].value_counts().plot(kind='bar')
```

```python
df.plot.scatter(x="prxval ()", y="phase (radians)", c="epc96", cmap="viridis", s=50, alpha=.3);
```

```python
a = df[df.epc96 == "e280116060000209e445f01a"]
b = df[df.epc96 == "e280116060000209e445ba0a"]
c = df[df.epc96 == "e280116060000209e4454c2a"]
d = df[df.epc96 == "e280116060000209e4454a0a"]
by_tag = [a, b, c, d]
for x in by_tag:
    x.plot.scatter(x="prxval ()", y="phase (radians)", c="antennastate", 
                   cmap="viridis", s=50, alpha=.3);
```
